#Introduction

The DE-SynPUF was created with the goal of providing a realistic set of claims data in the public domain while providing the very highest degree of protection to the Medicare beneficiaries’ protected health information.

Although the DE-SynPUF has very limited inferential research value to draw conclusions about Medicare beneficiaries due to the synthetic processes used to create the file, the Medicare DE-SynPUF does increase access to a realistic Medicare claims data file in a timely and less expensive manner to spur the innovation necessary to achieve the goals of better care for beneficiaries and improve the health of the population.

The DE-SynPUF contains five types of data – Beneficiary Summary, Inpatient Claims, Outpatient Claims, Carrier Claims, and Prescription Drug Events. More details about the data can be found [here](http://www.cms.gov/Research-Statistics-Data-and-Systems/Downloadable-Public-Use-Files/SynPUFs/DE_Syn_PUF.html).

#Downloading the Data

Due to file size limitations, each data type in the CMS Linkable 2008-2010 Medicare DE-SynPUF is released in 20 separate samples (essentially each is a .25% sample).  All claims for a particular beneficiary are in samples with the same number (i.e. all beneficiaries in sample 1 have all their claims in the sample 1 files).  This design allows DE-SynPUF users who do not need the entire synthetic population of the DE-SynPUF to read in only as many samples as they desire. You can download the data from [here](http://www.cms.gov/Research-Statistics-Data-and-Systems/Downloadable-Public-Use-Files/SynPUFs/DE_Syn_PUF.html).

#About Each Sample

Each of the 20 samples contains eight files – three beneficiary files (one for each year), one inpatient file containing three years of data, one outpatient file containing three years of data, one PDE file containing three years of data, and two carrier files containing three years of data (Carrier 1 and Carrier 2).  Because of file size limitations, a Carrier sample was split into two CSV files. Both CSV files in a sample must be downloaded. Beneficiary data was obtained for each year that the beneficiary enrolled in Medicare.  A single Beneficiary sample contains three CSV files, one for each year.  Because beneficiary files contain time varying variables like chronic conditions, reimbursement variables, and death, three files (one for each year) were provided to keep the same variable name as in the actual data.  All three CSV files in a sample must be downloaded. If the beneficiary dies between 2008 and 2010, the beneficiary will not have any data in years after the beneficiary’s death.

#PostgreSQL

If you plan on working with all the CMS data, I would recommend storing it in a database since it is quite large. To do this, I would recommend downloading [Postgres.app](http://postgresapp.com/). If you are not on Mac, check out this [article](https://www.codefellows.org/blog/three-battle-tested-ways-to-install-postgresql#windows). Once downloaded, you can then launch the app (or postgresql) and type psql in the terminal (this launches psql). Then:
* Create a CMS database:
    * CREATE DATABASE cms
* Now go to this database (\q out of psql first):
    * psql cms
* You now are in your cms database. You can now create a table (note: you can find the list of data types for PSQL [here](http://www.postgresql.org/docs/9.3/static/datatype.html)). Basically you create a table with all of the columns and data types. You can use the Cookbook to figure out what type of data are in each column of the CMS data.

```python
data_type = ['text', 'integer', 'integer', 'smallint', 'smallint',
             'character(1)', 'character(2)', 'text', 'smallint',
             'smallint', 'smallint', 'smallint', 'smallint',
             'smallint', 'smallint', 'smallint', 'smallint',
             'smallint', 'smallint', 'smallint', 'smallint',
             'smallint', 'smallint', 'real', 'real',
             'real', 'real', 'real', 'real',
             'real', 'real', 'real']
column_names = data.columns
with open("psql_table.txt", "w") as text_file:
    text_file.write("CREATE TABLE bsf_2008_1 (")
    for i in range(len(column_names)):
        if i == 0:  
            text_file.write(" {0} {1} ".format(column_names[i], 
                                         data_type[i]))
        else:
            text_file.write(", {0} {1} ".format(column_names[i], 
                                         data_type[i]))
    text_file.write(");")
```

The above code spits out a text file. You can then copy the output of this file and paste it into your terminal (still in the cms database). Now that you have an empty table you can copy your DE1_0_2008_Beneficiary_Summary_File_Sample_1.csv file into it. Here is the command for this:

\copy bsf_2008_1 FROM 'pathtodata/DE1_0_2008_Beneficiary_Summary_File_Sample_1.csv' DELIMITER ',' CSV HEADER;

Nice. Now your data are in PSQL. Note: If you mess up and need to delete the table and start over, the command is DROP TABLE tablename. Lets make sure our data match our csv data.

```python
import psycopg2
import pandas.io.sql as psql
db = psycopg2.connect("dbname='cms' host='localhost'")
psql_data = psql.read_sql("""select * from bsf_2008_1;""", con=db)
psql_data.head()
```

Looks promising. Lets check the shapes

```python
data.shape
```

```python
psql_data.shape
```

Cool. Now that we are in postgres we can query the data as opposed to reading it all into memory. Also, another useful thing is to create [indexes](http://stackoverflow.com/questions/1108/how-does-database-indexing-work) on our data. Indexes allow for faster searching. Imagine we are usually interested in searching for certain states. We could create an index on state like this:

CREATE INDEX on bsf_2008_1 (SP_STATE_CODE)

You can confirm the index was created by typing:

\d bsf_2008_1

More on PSQL indexes [here](http://www.postgresql.org/docs/9.1/static/sql-createindex.html)

Now you can search through states faster, so:

```python
state_26 = psql.read_sql(""" select * from bsf_2008_1 where SP_STATE_CODE = '26'; """, con=db);
state_26.sp_state_code.value_counts()
```

So, in this files there are 2378 observation with state id 26. Now maybe you want to see the distribution of inpatient annual medicare amount for this state. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks', palette='Set2')

state_26.medreimb_ip.plot(kind='hist')
plt.ylabel('')
sns.despine()
```

```python
state_26.bene_hmo_cvrage_tot_mons.value_counts()
```

It is pretty clear that people in this state either have HMO coverage for the entire year or don't have it at all. With most people not having it at all. How do inpatient amounts differ for HMO coverage?

```python
plt.scatter(state_26.bene_hmo_cvrage_tot_mons, state_26.medreimb_ip)
sns.despine()
```

Definitely looks like higher amounts paid without HMO coverage. Makes sense.
